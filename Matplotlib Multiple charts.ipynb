# Matplotlib Bar Chart

```python
classes = ['Python' , 'R' ,'Ai' , 'ML' , 'DS']
cls1_std = np.random.randint(20 , 100 , 5)
cls2_std = np.random.randint(10,90 , 5)
cls3_std = np.random.randint(30,80 , 5)
display(classes,cls1_std,cls2_std,cls3_std)
```

```python
plt.bar(classes,cls1_std)
plt.show()
```

```python
# plot the same bar horizantally
plt.barh(classes , cls1_std)
plt.show()
```

```python
plt.bar(classes , cls2_std , width=0.50)
plt.show()
```

```python
# to show bars on edge 
plt.bar(classes , cls2_std , width=0.50 ,align='edge')
plt.show()
```

```python
# to change color of the grph 
plt.bar(classes , cls2_std , width=0.5 ,align='center' , color = 'y' , edgecolor = 'r')
plt.show()
```

```python
# to change color of the grph 
# to expand the width of the edge lines
plt.bar(classes , cls2_std , width=0.5 ,align='center' , color = 'y' , edgecolor = 'r' , linewidth = 5)
plt.show()
```

```python
# to change color of the grph 
# to expand the width of the edge lines
# change opaciy of bar
plt.figure(figsize=(15,6))
plt.bar(classes , cls2_std , width=0.5 ,align='center' , color = 'y' , edgecolor = 'r' , linewidth = 5 , alpha = 0.5 ,
       linestyle = '--' , label = 'cls2 std')
plt.legend()
plt.show()
```

```python
# to change color of the grph 
# to expand the width of the edge lines
# change opaciy of bar
plt.figure(figsize=(15,6))
plt.bar(classes , cls2_std , width=0.5 ,color = 'y' , label = 'cls2 std')
plt.bar(classes , cls1_std , width=0.5 ,color = 'r' , label = 'cls1 std')
plt.legend()
plt.show()

# this is overlaping we use another method
```

```python
# to change color of the grph 
# to expand the width of the edge lines
# change opaciy of bar
class_index = np.arange(len(classes))
width = 0.2
plt.figure(figsize=(15,6))
plt.bar(class_index , cls2_std ,width ,color = 'y' , label = 'cls2 std')
plt.bar(class_index+width , cls1_std ,width , color = 'r' , label = 'cls1 std')
plt.legend()
plt.show()

# this is overlaping we use another method
```

```python
# to change color of the grph 
# to expand the width of the edge lines
# change opaciy of bar
plt.figure(figsize=(16,6))
class_index = np.arange(len(classes))
width = 0.2
plt.bar(class_index , cls1_std , width ,color = 'r')
plt.bar(class_index+width , cls2_std , width , color = 'y')
plt.bar(class_index+width+width , cls3_std , width , color = 'g')
plt.xticks(class_index+width , classes , rotation = 40)
plt.xlabel('Class index')
plt.ylabel('Profit')
plt.title('Profit by Class' , fontsize = 15)
# this is overlaping we use another method
```

# Matplotlib Scatter Plot

A scatter plot (aka scatter chart, scatter graph) uses dots to represent values for two different numeric variables. The position of each dot on the horizontal and vertical axis indicates values for an individual data point. Scatter plots are used to observe relationships between variables.

```python
df = pd.read_csv(r'https://raw.githubusercontent.com/attaurrahman294/Datasets/main/Fortune_10.csv')
df.shape
```

```python
# i am only getting 1000 rows from this data set
df = pd.read_csv(r'https://raw.githubusercontent.com/attaurrahman294/Datasets/main/Fortune_10.csv' , nrows=1000)
df.shape
```

```python
df.dtypes
```

```python
plt.scatter(df.Rating , df.Reviews)
plt.show()
```

```python
plt.figure(figsize=(16,7))
plt.scatter(df.Rating , df.Reviews )
plt.xlabel('Rating of the app')
plt.ylabel('Revies of the app')
plt.title('Google play store app' , fontsize = 20 , color = 'g')
```

```python
# change the marker shape
plt.figure(figsize=(16,7))
plt.scatter(df.Rating , df.Reviews , c='g' , marker='*')
plt.xlabel('Rating of the app')
plt.ylabel('Revies of the app')
plt.title('Google play store app' , fontsize = 20 , color = 'r')
```

```python
# extend size of the marker
plt.figure(figsize=(16,7))
plt.scatter(df.Rating , df.Reviews , c='g' , marker='*' ,s = 200)
plt.xlabel('Rating of the app')
plt.ylabel('Revies of the app')
plt.title('Google play store app' , fontsize = 20 , color = 'r')
```

```python
# extend size of the marker
# change tranferance
# change opacity
plt.figure(figsize=(16,7))
plt.scatter(df.Rating , df.Reviews , c='g' , marker='*' ,s = 200 , alpha=0.5)
plt.xlabel('Rating of the app')
plt.ylabel('Revies of the app')
plt.title('Google play store app' , fontsize = 20 , color = 'r')
```

```python
#show edge line of the marker
plt.figure(figsize=(16,7))
plt.scatter(df.Rating , df.Reviews , c='r' , marker='*' ,s = 200 , alpha=0.5 , linewidths=10 ,edgecolors='g' )
plt.xlabel('Rating of the app')
plt.ylabel('Revies of the app')
plt.title('Google play store app' , fontsize = 20 , color = 'r')
plt.show()
```

```python
# show two values on y label
plt.figure(figsize=(16,7))
plt.scatter(df.Price, df.Reviews , c='g' , marker='*' ,s = 200 , alpha=0.5)
plt.scatter(df.Price, df.Rating , c='r' , marker='*' ,s = 200 , alpha=0.5)
plt.xlabel('Rating of the app')
plt.ylabel('Revies of the app')
plt.title('Google play store app' , fontsize = 20 , color = 'r')
```

# Matplotlib Pie Char

A pie chart is a graphical representation technique that displays data in a circular-shaped graph. It is a composite static chart that works best with few variables. Pie charts are often used to represent sample data—with data points belonging to a combination of different categories. Each of these categories is represented as a “slice of the pie.” The size of each slice is directly proportional to the number of data points that belong to a particular category.

```python
display(classes , cls1_std , cls2_std,cls3_std)
```

```python
plt.pie(cls1_std , labels=classes )
plt.show()
```

```python
# explode some category for the circle
explode = [0.1,0,0.1,0.,0]
plt.pie(cls1_std , labels=classes , explode = explode)
plt.show()
```

```python
# change color of every category manuallyy
colors = ['r' , 'g' ,'b' ,'m','y']
plt.pie(cls1_std , labels=classes , colors=colors , autopct='%0.1f%%')
plt.show()
```

```python
# change color of every category manuallyy
colors = ['r' , 'g' ,'b' ,'m','y']
explode = [0.1 ,0 ,0,0,0]
plt.pie(cls1_std , labels=classes , explode = explode ,colors=colors , autopct='%0.4f%%')
plt.show()
```

```python
# change color of every category manuallyy
colors = ['r' , 'g' ,'b' ,'m','y']
plt.pie(cls1_std , labels=classes , colors=colors , autopct='%0.4f%%' , shadow=True)
plt.show()
```

```python
# to extend radious
colors = ['r' , 'g' ,'b' ,'m','y']
plt.pie(cls1_std , labels=classes , colors=colors , autopct='%0.4f%%' , shadow=True , radius=2.4)
plt.show()
```

```python
# start with manual angle
# to with any angle
# start with 180 angle
colors = ['r' , 'g' ,'b' ,'m','y']
plt.pie(cls1_std , labels=classes , colors=colors , autopct='%0.1f%%' , shadow=True , startangle=180)
plt.show()
```

```python
# to show percentange outside from the graph we can extend pct distance from o.6 to any number
colors = ['r' , 'g' ,'b' ,'m','y']
plt.pie(cls1_std , labels=classes , colors=colors , autopct='%0.1f%%' , shadow=True , pctdistance=1.6)
plt.show()
```

```python
# visevirsa play with label distance
colors = ['r' , 'g' ,'b' ,'m','y']
plt.pie(cls1_std , labels=classes , colors=colors , autopct='%0.1f%%' , shadow=True , labeldistance=1.9)
plt.show()
```
